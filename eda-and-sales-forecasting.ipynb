# Relationships between variables

```python
# Calculating log of sales
df_train['Log_of_Sales'] = [np.log(s) for s in df_train['Sales']]
df_train.describe()
```

```python
sns.set_theme(style='ticks', color_codes=True)
sns.catplot(data=df_train, x='Log_of_Sales', y='Segment', hue='Category', kind='box', row='Region', orient='h', aspect=1.5, height=5)

plt.figure(figsize = (15,8))
```

# Time Series Plots

* Plotting sales over time is not too helpful.

```python
df_train_order = df_train[['Order Date', 'Sales']]
df_train_order.plot('Order Date', 'Sales')
plt.figure(figsize=(10, 6), dpi=80)
plt.show()
```

```python
df_train_order.hist('Order Date', bins=20)
binwidth = 2
df_train_order.hist('Sales', bins=range(int(min(df_train_order['Sales'])), 150, binwidth))
plt.show()
```

It is too diffuicult to see any trends so I'll calculate a moving average

```python
df_train_rolling = df_train[['Order Date', 'Sales']]
df_train_rolling['Sales'] = df_train_rolling['Sales'].rolling(300).sum()
print(df_train_rolling.head(40))
```

```python
df_train_rolling.plot('Order Date', 'Sales')
plt.title('Rolling Average')
plt.show()
```
